```python
from langchain.messages import HumanMessage
import json
from langgraph_graph import graph, collector
```

```python
collector.reset()

# Run graph
result =  graph.invoke({"messages": [HumanMessage(content="what is 2 multiplied by 5?")]})
print(result["messages"][-1].content)

print("-- Final Trace JSON --")

trace_json = collector.build()
print(json.dumps(trace_json, indent=4))
```

```python
collector.reset()

# Run graph
for node in graph.stream({"messages": [HumanMessage(content="what is 2 multiplied by 5?")]}):
    print(node)

print("-- Final Trace JSON --")

trace_json = collector.build()
print(json.dumps(trace_json, indent=4))
```
